# Data simulation
Use preprocessed ECOSTRESS hyperspectral data to simulate the reflectance of Sentinel-2 A/B, B2, B3, and B4 bands, and also calculate CIE XYZ values.

Spectral range: 360nm to 830nm

# Sentinel-2 band reflectance calculation method

A Lamterian type surface sample $j$, with a reflectance spectrum $\rho_{j}(\lambda)$, illuminated by an irradiance spectrum $E(\lambda)$, has a reflected radiance spectrum given by:
$$L_{j}(\lambda) = \frac{1}{\pi} E(\lambda) \cdot \rho_{j}(\lambda)$$

This radiance spectrum will be measured by the spectral band $i$ of an optical sensor, whose spectral response function SRF is $S_{i}(\lambda)$. The band-integrated radiance $L_{i,j}$ is given by:
$$L_{i,j} = \frac{ \int_{0}^{+\infty} L_{j}(\lambda)S_{i}(\lambda)d\lambda }{ \int_{0}^{+\infty} S_{i}(\lambda)d\lambda } = \frac{ \frac{1}{\pi} \int_{0}^{+\infty} E(\lambda)\rho_{j}(\lambda)S_{i}(\lambda)d\lambda }{ \int_{0}^{+\infty} S_{i}(\lambda)d\lambda } $$ 

and the band reflectance $\rho_{i,j}$ is given by:
$$\rho_{i,j} = \frac{\pi \cdot L_{i,j}}{E_{i}} = \frac{\int_{0}^{+\infty} E(\lambda)\rho_{j}(\lambda)S_{i}(\lambda)d\lambda}{\int_{0}^{+\infty} E(\lambda)S_{i}(\lambda)d\lambda}$$


We use the ECOSTRESS library ground object surface reflectance spectrum as $\rho_{j}(\lambda)$ and the CIE D65 light source irradiance spectrum as $E(\lambda)$.

We calculate the equivalent Sentinel-2 A/B ground reflectance of bands B2, B3, and B4. 

# CIE1931 XYZ calculation method

$$X_{j} = \frac{1}{N} \int E(\lambda) \rho_{j}(\lambda) \bar x(\lambda) d\lambda$$
$$Y_{j} = \frac{1}{N} \int E(\lambda) \rho_{j}(\lambda) \bar y(\lambda) d\lambda$$
$$Z_{j} = \frac{1}{N} \int E(\lambda) \rho_{j}(\lambda) \bar z(\lambda) d\lambda$$
where $E(\lambda)$ is the CIE D65 irradiance sectrum, $rho_{j}(\lambda)$is the ECOSTRESS reflectance spctrum, $\bar x(\lambda),\bar y\lambda),\bar z(\lambda)$ are the three normalized x, y, and z color spectrum distribution curves and $N$ is a normalizing constant.

$$N=\int E(\lambda) \bar y(\lambda) d\lambda$$



```python
%load_ext autoreload
%autoreload 2
import os 
import glob
import numpy as np
# import matplotlib.pyplot as plt
import sys

current_dir = os.getcwd()
sys.path.append(os.path.abspath(os.path.join(current_dir,'..')))
from src import *
```

```python
# ECOSSTRESS spec lib
spec_srcpath = os.path.join(current_dir,'ECOSTRESS')
srcflist = np.array(glob.glob(os.path.join(spec_srcpath,'*.spectrum.csv')))
splib_names = [f[len(spec_srcpath)+1:-len('.csv')] for f in srcflist]
print('Total number of ECOSTRESS spectra',len(splib_names))

# D65 spec
ds_d65 = read_csv(os.path.join(current_dir,'CIE','CIE_std_illum_D65.csv'))
wls = np.array(ds_d65['WL'])
d65 = np.array(ds_d65['D65'])

# Sentinel-2 SRF
ds_srf_s2a = read_csv(os.path.join(current_dir,'SRF','S2A-SRF.csv'))
ds_srf_s2b = read_csv(os.path.join(current_dir,'SRF','S2B-SRF.csv'))
s2_bnames = ['B2','B3','B4']
sensor_names = ['S2A','S2B']
ds_srf_s2 = [ds_srf_s2a,ds_srf_s2b]

# load xyz
ds_xyz = read_csv(os.path.join(current_dir,'CIE','cie_1931_2deg_xyz_1nm.csv'))
```

```python
def band_resp(spec,srf):
    val = np.sum(spec*srf)/np.sum(srf)
    return val
```

```python
# Calculate the constants of D65 of three S2 bands
def simu_sentinel2_d65(ds_srf,s2_bnames,d65,dstfile):
    mydict = {}
    mydict['BNAME'] = s2_bnames
    vals = [] 
    for i in range(len(s2_bnames)):
        bname = s2_bnames[i]
        srf = np.array(ds_srf[bname])
        val = band_resp(d65,srf)
        vals.append(val)
    mydict['D65'] = np.array(vals)
    write_csv(dstfile,mydict)
    
for i in range(len(ds_srf_s2)):
    simu_sentinel2_d65(ds_srf_s2[i],s2_bnames,d65,os.path.join(current_dir,'D65_'+sensor_names[i]+'.csv'))
```

```python
# Sentinel-2 A/B simulation
def simu_sentinel2_refl(ds_srf,s2_bnames,d65,srcflist,splib_names,dstfile):
    mydict = {}
    mydict['NAME'] = splib_names
    for i in range(len(s2_bnames)):
        bname = s2_bnames[i]
        srf = np.array(ds_srf[bname])
        vals = []    
        for j in range(len(srcflist)):
            srcfile = srcflist[j]
            ds_spec = read_csv(srcfile)
            spec = np.array(ds_spec['value'])
            # rho
            val = band_resp(spec,d65*srf)
            vals.append(val)
        mydict[bname] = np.array(vals)
    write_csv(dstfile,mydict)

def simu_sentinel2_rad(ds_srf,s2_bnames,d65,srcflist,splib_names,dstfile):
    mydict = {}
    mydict['NAME'] = splib_names
    for i in range(len(s2_bnames)):
        bname = s2_bnames[i]
        srf = np.array(ds_srf[bname])
        vals = []    
        for j in range(len(srcflist)):
            srcfile = srcflist[j]
            ds_spec = read_csv(srcfile)
            spec = np.array(ds_spec['value'])
            # rad
            val = band_resp(spec*d65,srf)
            vals.append(val)
        mydict[bname] = np.array(vals)
    write_csv(dstfile,mydict)
    

for i in range(len(ds_srf_s2)):
    simu_sentinel2_refl(ds_srf_s2[i],s2_bnames,d65,srcflist,splib_names,os.path.join(current_dir,'rho_ECOSTRESS_'+sensor_names[i]+'.csv'))
    simu_sentinel2_rad(ds_srf_s2[i],s2_bnames,d65,srcflist,splib_names,os.path.join(current_dir,'rad_ECOSTRESS_'+sensor_names[i]+'.csv'))
```

```python
# XYZ
def simu_xyz(ds_xyz,d65,srcflist,splib_names,dstfile):
    my_dict = {}
    my_dict['NAME'] = splib_names
    xyz_names = list(ds_xyz.keys())[1:]
    # the normalized constant N
    srf = np.array(ds_xyz['Y'])
    N = np.sum(d65*srf)
    for i in range(len(xyz_names)):
        name = xyz_names[i]
        srf = np.array(ds_xyz[name])
        vals = []    
        for j in range(len(srcflist)):
            srcfile = srcflist[j]
            ds_spec = read_csv(srcfile)
            spec = np.array(ds_spec['value'])
            val = np.sum(spec*d65*srf)/N          
            vals.append(val)
        my_dict[name] = np.array(vals)
    write_csv(dstfile,my_dict)
 
simu_xyz(ds_xyz,d65,srcflist,splib_names,os.path.join(current_dir,'xyz_ECOSTRESS.csv'))
```
