```python
import matplotlib.pyplot as plt
import numpy as np
import os
from PIL import Image
```

```python
chemin = [os.path.abspath("../Sentinel-2_to_XYZ/scat_delta_Lab_Eq8_new.png"), os.path.abspath("../Sentinel-2_to_XYZ/scat_delta_Lab_Eq8.png"),os.path.abspath("../Sentinel-2_to_XYZ/scat_delta_Lab_Eq9.png")]

def recup(chemin):
    images = []
    for k in range(len(chemin)):
        img = Image.open(chemin[k])
        images.append(img)
    return images

images = recup(chemin)
fig, axs = plt.subplots(3, 1, figsize=(15, 15))

axs[0].imshow(images[0], cmap='gray')
axs[0].axis('off') 
axs[0].set_title('Eq8_new')

axs[1].imshow(images[1], cmap='gray')
axs[1].axis('off')
axs[1].set_title('Eq8')


axs[2].imshow(images[2], cmap='gray')
axs[2].axis('off')
axs[2].set_title('Eq9')

plt.tight_layout()
plt.show()

print('a : Dimension de la couleur allant du vert (valeurs nÃ©gatives) au rouge (valeurs positives).')
print('b : Dimension de la couleur allant du bleu (valeurs nÃ©gatives) au jaune (valeurs positives).')
print('On remarque que dans l Ã©quation 8 qu il a un peu de variation niveau du vert et beaucoup au niveau du jaune')
print('Tandis que l Ã©quation 9 reflÃ¨te bcp dans le vert et rouge et jaune')


```

```python
# Chemin vers le dossier contenant les images
image_folder = os.path.abspath("../image/Amposta(Spain)/true_color")


# Liste pour stocker les images
images = []

# RÃ©cupÃ©rer toutes les images du dossier
for filename in os.listdir(image_folder):
    if filename.endswith('.tiff'):
        # Charger l'image
        img_path = os.path.join(image_folder, filename)
        img = Image.open(img_path)
        images.append(img)
        


# CrÃ©er une figure avec une grille de 2x2 pour afficher 4 images (2 en haut, 2 en bas)
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Afficher chaque image
axs[0, 0].imshow(images[0], cmap='gray')
axs[0, 0].set_title('Amposta_LC1_direct')
axs[0, 0].axis('off')  # Supprimer les axes

axs[0, 1].imshow(images[1], cmap='gray')
axs[0, 1].set_title('Amposta_LC1_T8_new')
axs[0, 1].axis('off')

axs[1, 0].imshow(images[2], cmap='gray')
axs[1, 0].set_title('Amposta_LC1_T8')
axs[1, 0].axis('off')

axs[1, 1].imshow(images[3], cmap='gray')
axs[1, 1].set_title('Amposta_LC1_T9')
axs[1, 1].axis('off')

# Ajuster la mise en page pour que les sous-plots soient bien espacÃ©s
plt.tight_layout()
plt.show()


```
